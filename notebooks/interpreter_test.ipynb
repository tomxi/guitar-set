```python
from guitar_set import interpreter as itp
import matplotlib.pyplot as plt
import os
import jams

%matplotlib inline


def get_all_jamses(a_dir, ext='.jams'):
    return [os.path.join(a_dir, name) for name in os.listdir(a_dir)
            if name.endswith(ext)]
```

```python
base_dir = '/Users/tom/Music/DataSet/guitar_set/vl'
jams_list = get_all_jamses(base_dir, 'hexc.jams')
for jam_path in jams_list:
    jam = jams.load(jam_path)
    out_path = os.path.join(base_dir, os.path.split(jam_path)[1].split('.')[0] + '_syn.wav')
    itp.sonify_jams(jam, out_path, q=0)
    itp.visualize_jams(jam)
```
