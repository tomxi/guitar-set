```python
import guitar_set.interpreter as itp
import jams
import os
import glob
```

```python
jams_list
```

```python
for j in jams_list:
    jam = jams.load(j)
    itp.visualize_jams(jam)
    base_name_list = os.path.basename(j).split('_')
    base_name = '_'.join(base_name_list[0:3])
    sonify_outpath = os.path.join(folder, 'sonify/{}_syn.wav'.format(base_name))
    try:
        itp.sonify_jams(jam, sonify_outpath)
    except:
        pass
```
