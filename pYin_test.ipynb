```python
from guitar_set import annotator as ghex
import librosa
import vamp
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = '/Users/tom/Music/DataSet/guitar_set/js_post_mira/js_Rock1-130-A_c/01_A_ghex_mira_mode-1_preset-js_Rock1-130-A_c.wav'
y, fs = librosa.load(f, sr=44100)

param = {
            'threshdistr': 2,
            'outputunvoiced': 0,
            'precisetime': 0,
            'lowampsuppression': 0.01,
            'onsetsensitivity': 0.3
        }

output_notes = vamp.collect(y, fs, 'pyin:pyin', output='notes', parameters=param)
output_sal = vamp.collect(y, fs, 'pyin:pyin', output='candidatesalience', parameters=param)
print(len(output_notes['list']))
```

```python
stepsize, sal_data = output_sal["matrix"]
my_plt = plt.imshow(sal_data[400:800].T[200:300])
plt.colorbar()
fs = 44100

def hz_to_bin(hz):
    return (fs/hz)
```

```python
for note in output_notes['list']:
    hz = note['values'][0]
    midi_pitch = librosa.hz_to_midi(hz)[0]
    bin_num = int(round(hz_to_bin(hz)))
    st = note['timestamp']
    et = st + note['duration']
    print(hz, midi_pitch, st, et)
    # start frame, end frame
    sf = int(round(float(st) / float(stepsize)))
    ef = int(round(float(et) / float(stepsize)))
    time_sliced = sal_data[sf:ef].T
    
    half_len = int(round((sf-ef)/2))
    note_region = time_sliced[bin_num-10:bin_num+10]
    avg_note_sal = np.mean(np.sum(note_region,0))
    print(avg_note_sal)
    my_plt = plt.imshow(time_sliced[bin_num-10:bin_num+10])
    
#     aggregate_t = np.sum(time_sliced,1) / (ef-sf)
#     print((aggregate_t.max(),aggregate_t.sum()))
#     my_plt = plt.plot(aggregate_t)
    plt.show()
    
    
    

```
